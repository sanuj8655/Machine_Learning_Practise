## Single Linear Regression

### import required package

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### read data from source and describing

```python
df = pd.read_csv('hearing_test.csv')
df.head()
```

```python
print(df.columns)
```

```python
print(df.describe())
```

```python
## chec for data types
print(df.info())
```

```python
sns.pairplot(df)
```

### check the relation between variables

```python
corr = df.corr()
corr.style.background_gradient(cmap = 'Greens_r')
```

### select input and op variable

```python
x = df.drop('test_result', axis=1)
y = df['test_result']
```

### split the data


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state= 12345, train_size= 0.8)
```

### creating a model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
## fir the data
model.fit(x_train, y_train)
```

### parameters to fine tune model

```python
print(model.get_params())
```

### predict the values for unseen data

```python
y_prediction = model.predict(x_test)
#print(y_prediction)
```

### evaluation of classification model

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_prediction)
print(cm)
```

```python
correct = cm[0,0] + cm[1,1]
wrong = cm[1,0] + cm[0,1]
total = correct + wrong
accuracy = correct/total
print(accuracy)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_prediction))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_prediction))
```
